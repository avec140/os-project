```python
from urllib.request import urlopen
html=urlopen(
    "https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EC%BD%94%EB%A1%9C%EB%82%98"
).read().decode('utf-8')
print(html)
```

```python
from bs4 import BeautifulSoup
naver=BeautifulSoup(html,features='lxml')
# text=naver.find('div',{'class','news_area'})
text=naver.find_all('a',{'class','news_tit'})
for m in text:
    print(m.get_text())
```
